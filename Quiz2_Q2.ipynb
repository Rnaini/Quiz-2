```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```
# Data cache
df = pd.read_csv("/content/dataset_31_credit-g-preprocessed.csv") 
df.head()
```

```
x,label = make_moons(n_samples=500, noise=0.2)
x[0:6]
```

```
label
```

```
model = DBSCAN(eps= 0.2, min_samples=10)
model.fit(x)
```

```
model.labels_
```

```
#Plot for predicted model
fig, ax = plt.subplots(figsize= (10,5))
ax.scatter(x[:, 0], x[:, 1], c=model.labels_)
fig.show()

#plot for original model
fig, ax = plt.subplots(figsize= (10,5))
ax.scatter(x[:, 0], x[:, 1], c=label)
fig.show()
```

```
from sklearn.metrics.cluster import homogeneity_score
homogeneity_score(label,model.labels_)
```

```
from sklearn.metrics.cluster import completeness_score
completeness_score(label,model.labels_)
```

```
from sklearn.metrics.cluster import v_measure_score
v_measure_score(label,model.labels_)
```
